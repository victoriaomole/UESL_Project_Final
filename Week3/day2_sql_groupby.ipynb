```python
# ðŸŸ¦ Week 3 â€” Day 2: SQL GROUP BY
import sqlite3
import pandas as pd

conn = sqlite3.connect("day2_week3_students.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    department TEXT,
    score INTEGER
);
""")

cursor.executemany("INSERT INTO students (name, department, score) VALUES (?, ?, ?)", [
    ("Victoria", "Data", 95),
    ("Daniel", "Data", 88),
    ("Mariam", "Finance", 72),
    ("James", "HR", 81),
    ("John", "Finance", 90)
])
conn.commit()

print("ðŸŽ“ All Students:")
print(pd.read_sql_query("SELECT * FROM students;", conn))

# Group by department and average score
query = """
SELECT department, COUNT(*) AS total_students, AVG(score) AS avg_score
FROM students
GROUP BY department;
"""
print("\nðŸ“Š Average Score by Department:")
print(pd.read_sql_query(query, conn))

conn.close()
```
